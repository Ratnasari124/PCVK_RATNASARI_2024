4. Membuat bentuk Geometri 2D dari OpenCV. Diawali dengan pembuatan black image dengan tipe data int16.

```python
black_img = np.zeros(shape=(512, 512, 3), dtype=np.int16)
plt.imshow(black_img)
```

Kemudian menambahkan bentuk persegi panjang sesuai koordinat pt1 dan pt2

```python
#
cv.rectangle(black_img,pt1=(384,0),pt2=(510,150),color=(0,255,0),thickness=10)
plt.imshow(black_img)
```

Selanjutnya ditambah menambahkan bentuk persegi sesuai koordinat pt1 dan pt2
yang tertulis pada kode program

```python
cv.rectangle(black_img,pt1=(200,200),pt2=(300,300),color=(0,0,255),thickness=15)
plt.imshow(black_img)
```

Tahap selanjutnya ditambah menambahkan bentuk lingkaran sesuai radius yang
tertulis pada kode program.

```python
cv.circle(black_img,center=(100,100),radius=50,color=(255,0,0),thickness=8)
plt.imshow(black_img)
```

Tahap selanjutnya ditambah menambahkan bentuk lingkaran penuh

```python
cv.circle(black_img,center=(400,400),radius=50,color=(255,0,0),thickness=-1)
plt.imshow(black_img)
```

Kemudian dilakukan penambahan garis sesuai koordinat pt1 dan pt2 sebagai berikut.

```python
cv.line(black_img, pt1=(0,0), pt2=(512,512),color=(255,255,255),thickness=5)
plt.imshow(black_img)
```

Penambahan text dengan font yang telah tertulis dengan ukuran yang sudah
ditentukan.

```python
font = cv.FONT_HERSHEY_SIMPLEX
cv.putText(black_img, text='Hello', org=(10,500), fontFace=font, fontScale=4,color=(255,255,0),thickness=2,lineType=cv.LINE_AA)
plt.imshow(black_img)
```

Pembuatan black image kembali dilakukan dengan tipe data int32

```python
black_img2=np.zeros(shape=(512,512,3),dtype=np.int32)
plt.imshow(black_img2)
```

Berikut adalah kode program untuk inisialisasi NumPy array dengan tipe data int32

```python
vertices = np.array([[100,300],[200,200],[400,300],[200,400]],dtype=np.int32)
vertices
```

Array tersebut kemudian di reshape sebagai berikut

```python
pts = vertices.reshape((-1,1,2)) #nilai 2 untuk menunjukkan bahwa tiap titik dibuat 3 channel yg mewakili R, G, dan B
pts
```

Penambahan polyline pada black image kedua yang telah dibuat.

```python
cv.polylines(black_img2,[pts],isClosed=True,color=(255,0,0),thickness=5)
plt.imshow(black_img2)
```

1. Apakah perbedaan gambar yang ditampilkan tanpa dan dengan matplotlib?

Jawaban :   
*   Tanpa Matplotlib

Menggunakan OpenCV dengan Function cv.imshow(). Gambar ditampilkan dalam format BGR (Blue, Green, Red), karena OpenCV menggunakan format ini secara default. hal ini menyebabkan kurang fleksibel untuk pengaturan tambahan misal untuk menambah grid. dan gambar akan ditampilkan dalam jendela GUI terpisah yg dibuat oleh OpenCV, yang menjadikan gambar nya dapat di perbesar atau diperkecil menggunakan mouse. jadi OpenCV lebih cocok untuk tampilan langsung dengan interaksi pengguna minimal


*   Dengan Matplotlib

Menggunakan Matplotlib dengan fuction plt.imshow(). format gambarnya adalah RGB (Red, Green, Blue). jadi gambar yang dihasilkan oleh OpenCV (BGR) harus dikonversi ke RGB sebelum ditampilkan.Dengan Matplotlib, gambar ditampilkan dalam plot yang memungkinkan penambahan sumbu, anotasi, dan kustomisasi lainnya, sehingga cocok untuk menampilkan beberapa gambar atau hasil analisis dalam satu figure. sehingga Matplotlib memberikan lebih banyak kontrol visual dan fleksibilitas dalam penyesuaian gambar.

Jawaban :    
*   int16

setiap komponen warna (merah, hijau, biru) bisa memiliki nilai antara -32768 hingga 32767. kapasitas penyimpannaya lebih kecil.

*   int32

komponen warna bisa memiliki nilai antara -2147483648 hingga 2147483647.memori yang digunakan 2 kali lipat dari pada int16, sehingga nilai pixelnya juga lebih besar tanpa adanya resiko overflow.

```python
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
import matplotlib.pyplot as plt
import numpy as np
img_tugas = cv.imread('/content/drive/MyDrive/ratna/IMG_1716.JPG')
plt.imshow(img_tugas)

tinggi = img_tugas.shape[0]
lebar = img_tugas.shape[1]
print("resolusi image:tinggi x lebar = ", tinggi, " x ", lebar)
cv2_imshow(img_tugas)
```

figsize hanya mengubah ukuran tampilan gambar dalam plot Matplotlib, bukan ukuran piksel gambar itu sendiri. Ukuran gambar dalam piksel tetap sama.

```python
img_t = img_tugas
img_t2 = cv.cvtColor(img_tugas, cv.COLOR_BGR2RGB) #konversi channel BGR -> RGB
plt.imshow(img_t2)
```

```python
# Red-Blue channel: Set Green channel to 255 (putih)
img_tugas = img_tugas.copy()
img_tugas[:, :, 1] = 255  # Set Green channel to 255 (putih)

# Green-Blue channel: Set Red channel to 255 (putih)
img_tugas = img_tugas.copy()
img_tugas[:, :, 2] = 255  # Set Red channel to 255 (putih)

# Menampilkan gambar dalam channel Red-Blue
plt.subplot(1, 2, 1)
plt.imshow(cv.cvtColor(img_tugas, cv.COLOR_BGR2RGB))
plt.title('Red-Blue Channel')
plt.axis('off')

# Menampilkan gambar dalam channel Green-Blue
plt.subplot(1, 2, 2)
plt.imshow(cv.cvtColor(img_tugas, cv.COLOR_BGR2RGB))
plt.title('Green-Blue Channel')
plt.axis('off')

plt.show()
```
