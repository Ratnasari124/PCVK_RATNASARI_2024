```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow # for image display
from skimage import io
from skimage import transform
from PIL import Image
import matplotlib.pylab as plt
```

```python
# membat list untuk menyimpan url dar beberapa image
urls = ["https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg",
        "https://iiif.lib.ncsu.edu/iiif/0016007/full/800,/0/default.jpg",
        "https://air.eng.ui.ac.id/images/a/a7/Kucing.jpg"]
        #"https://asset-2.tstatic.net/pontianak/foto/berita/2012/5/16/SIM_C.jpg",]

# baca dan tampilkan image
# loop pada tiap url image, beberapa image dapat disimpan pada list

for url in urls:
  image = io.imread(url)   #read image
  image = cv.resize(image, (0,0), fx=0.5, fy=0.5) #resize image to half size
  image_2 =cv.cvtColor(image, cv.COLOR_BGR2RGB) #convert coclor to rgb
  final_frame  = cv.hconcat((image, image_2)) #concatenate image
  cv2_imshow(final_frame) #concatenate image
  print('\n')
```

```python
import cv2 as cv
import matplotlib.pyplot as plt
import random

# Akses file image lokal (KTP/SIM) dan tampilkan menggunakan OpenCV dan Matplotlib
image_path = '/content/sample_data/gambar ktp.jpg'
image_satu = cv.imread(image_path)
image_satu = cv.resize(image_satu, (0,0), fx=0.5, fy=0.5) #ubah ukuran
image_rgb = cv.cvtColor(image_satu, cv.COLOR_BGR2RGB)  # Konversi untuk Matplotlib

# Tampilkan menggunakan OpenCV
print("Gambar KTP (Tampilan Menggunakan OpenCV)")
#cv.imshow('Gambar KTP/SIM (OpenCV)', image_satu) #formatnya sebenarnya ini
cv2_imshow(image_satu)
cv.waitKey(0)
cv.destroyAllWindows()

# Tampilkan menggunakan Matplotlib
plt.title('Gambar KTP (Tampilan Menggunakan Matplotlib)')
plt.imshow(image_rgb)
plt.axis('off')
plt.show()

# Fungsi untuk menutupi bagian tertentu
def tutup_bagian(image, x_awal, y_awal, lebar, tinggi, warna):
    for y in range(y_awal, y_awal + tinggi):
        for x in range(x_awal, x_awal + lebar):
            image[y, x] = warna

# Tentukan kelompok
kelompok = 1

if kelompok == 1:
    # Menutupi bagian NIK dan Nama
    tutup_bagian(image_rgb, 120, 55, 230, 20, [0, 255, 0])  # Kotak 1
    tutup_bagian(image_rgb, 150, 85, 200, 15, [255, 0, 0])  # Kotak 2
elif kelompok == 2:
    # Menutupi bagian TTL dan Jenis Kelamin
    tutup_bagian(image_rgb, 150, 85, 200, 15, [0, 255, 0])  # Kotak 1
    tutup_bagian(image_rgb, 150, 85, 200, 15, [255, 0, 0])  # Kotak 2
# kalo mau tambah ketentuan lain bisa di add di sini

# Tampilkan hasil modifikasi dengan Matplotlib
plt.imshow(image_rgb)
plt.title(f'Gambar KTP setelah ditutupi (Kelompok {kelompok})')
plt.axis('off')
plt.show()
```
